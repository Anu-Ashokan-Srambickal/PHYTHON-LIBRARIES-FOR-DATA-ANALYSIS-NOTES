### ‚òòÔ∏è LINE PLOT

- plt.plot : is used to create a line plot
- x axis and y axis are mandatory for line plot
- memory details of graph can be removed by either plt.show() or ;
- Data point can be specified by marker using o,*,+  etc.
- Colour also can be customized
- linestyle'--', linewidth,

```python
quaters = [1, 2,3,4]
revenue = [100, 200,90, 180]
```

```python

plt.plot(quaters, revenue, marker='o', color='k', linestyle='--',linewidth=5)
plt.xlabel('Quaters', size=15)
plt.ylabel('Revenue', size=15)
plt.title('Revenue Generated Per Quater', size=20)
plt.show()
```

#### üî∞ MULTIPLE Y AXIS VALUES IN LINE PLOT - plt.plot

```python
quaters = [1, 2,3,4]
revenue_2017 = [100, 200,90, 180]
revenue_2018 = [90, 250, 190, 250]
```

```python
plt.plot(quaters, revenue_2017, marker='o', color='k', linestyle='--',linewidth=5)
plt.plot(quaters, revenue_2018, marker='*', color='y',linewidth=2)
plt.xlabel('Quaters', size=15)
plt.ylabel('Revenue', size=15)
plt.title('Revenue Generated Per Quater', size=20)
plt.legend(['Sales in 2017', 'Sales in 2018'])
plt.show()
```

### ‚òòÔ∏è BAR CHART - plt.bar

```python
Categories = ['Categ_A', 'Categ_B', 'Categ_C', 'Categ_D']
Value_One = [15, 24, 30, 18]
Value_Two = [8,16,10,22]
```

### ‚òòÔ∏è PIE CHART - plt.pie

- labels is used for labeling
- autopct='%1.1f%%' is a code to add percentage in pie chart

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 24, 30, 18]
```

```python
plt.pie(sizes, labels = labels, autopct='%1.1f%%');
```

```python
plt.pie(sizes, labels = labels, autopct='%1.1f%%',colors=['skyblue', 'lightcoral', 'lightgreen'],
        wedgeprops={'edgecolor': 'white', 'linewidth': 1.5});
```
