#### üçí PIVOT TABLES

- Creating a pivot table in Python is super straightforward when using the pandas library.
- It‚Äôs a powerful way to summarize and analyze data


```python
data = {
    'Region': ['East', 'West', 'East', 'North', 'West', 'South'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'C'],
    'Sales': [100, 200, 150, 300, 250, 400]
}

df = pd.DataFrame(data)
```

```python
pivot = pd.pivot_table(df, 
                       index='Region',     # rows
                       columns='Product',  # columns
                       values='Sales',     # values to aggregate
                       aggfunc='sum',      # aggregation function
                       fill_value=0)       # fill missing values
```

#### üß† Bonus Tips

- You can use other aggregation functions like 'mean', 'count', 'max', 'min', or even custom functions.
- For multi-level pivot tables, you can pass lists to index or columns.
