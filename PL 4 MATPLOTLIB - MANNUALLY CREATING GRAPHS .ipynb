### üçí CREATING A DIFFERENT SUB PLOT 

```python
x = [100, 200, 300, 400]
y = [1250, 780, 810, 1875]
```

```python
a = np.arange(0,10)
b = (a**5) + 25
```

```python
fig = plt.figure(figsize=(20,8), dpi =250)


ax = fig.add_axes([0, 0, 1, 1])
ax1 = fig.add_axes([0.4, 0.5, 0.3, 0.3])

ax.plot(x,y, color='blue')
ax1.plot(a,b, color='red');
```

#### ‚òòÔ∏è LABELING X AND Y AXIS  - ax.set_xlabel & ylabel

```python
fig = plt.figure(figsize=(20,8), dpi =250)


ax = fig.add_axes([0, 0, 1, 1])
ax1 = fig.add_axes([0.4, 0.5, 0.3, 0.3])

ax.plot(x,y, color='blue')
# labeling
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

ax1.plot(a,b, color='red');
ax1.set_xlabel('A axis')
ax1.set_ylabel('B axis')
```
